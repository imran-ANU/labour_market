```python
import geopandas
```

```python
data_collect = geopandas.datasets.get_path("nybb")
```

```python
gdf = geopandas.read_file(data_collect)
```

```python
gdf
```

```python
gdf.to_file("my_file.geojson", driver="GeoJSON")
```

```python
gdf["area"] = gdf.area
gdf["area"]
```

```python
gdf['Shape_Area']
```

```python
gdf.boundary
```

```python
gdf['boundary'] = gdf.boundary
gdf['boundary']
```

```python
gdf['centroid'] = gdf.centroid
gdf['centroid']
```

```python
first_point = gdf.centroid.iloc[0]  #first point of the centroid
```

```python
gdf['distance'] = gdf.centroid.distance(first_point)
gdf['distance']
```
